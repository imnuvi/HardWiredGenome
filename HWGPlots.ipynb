```python
from constants import HARD_WIRED_GENOME_A_MATRIX_PATH
import pandas as pd
```

```python
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
```

```python
A_matrix = pd.read_csv(HARD_WIRED_GENOME_A_MATRIX_PATH, index_col = 0)
```

```python
sparse_adj = csr_matrix(A_matrix.values)
```

```python
plt.figure(figsize=(10, 10))
plt.spy(sparse_adj, markersize=0.5)
plt.title("A matrix from HURI + STRING")
plt.xlabel("Genes")
plt.ylabel("Genes")
plt.tight_layout()
# plt.show()
plt.savefig('figures/A_matrix.png', dpi=300)
```
